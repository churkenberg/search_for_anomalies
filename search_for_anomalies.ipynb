```python
import pandas as pd
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

```python
random_values = [random.randint(1, 100) for _ in range(100)]
random_values.append(200)
random_values.append(250)
random_values.append(300)
random_values.append(270)
random_values.append(280)
random_values.append(240)
data = pd.DataFrame(random_values)
```

```python
def search_anomalies(data):

  sns.boxplot(data, color='skyblue')
  plt.title('data')
  plt.show()

  print('\n \n \n')

  # Define number of trees in the forest
  n_estimators = len(data)

  # Define the model
  model = IsolationForest(n_estimators=n_estimators)

  #Fit the model to your data
  model.fit(data)

  #Get the anomaly scores
  anomaly_scores = model.decision_function(data)

  sns.scatterplot(anomaly_scores, c=anomaly_scores, cmap='plasma')
  plt.xlabel('index')
  plt.ylabel('scores')
  plt.title('anomaly scores')
  plt.show()

  print('\n \n \n')

  print(f'array size: {n_estimators} \n')
  print(anomaly_scores, '\n')

  for i in range(n_estimators):
    anomaly_scores[i]
    anomaly_scores_series = pd.Series(anomaly_scores)
    q = anomaly_scores_series.quantile(0.95)
    if (anomaly_scores[i] > q):
      print(f'outlier ({i}): {anomaly_scores[i]}')
```

```python
search_anomalies(data)
```
